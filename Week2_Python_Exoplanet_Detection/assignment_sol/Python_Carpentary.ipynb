```python
import numpy
import matplotlib.pyplot

data = z

fig = matplotlib.pyplot.figure(figsize=(16.0, 5.0))

axes2 = fig.add_subplot(5, 1, 1)
axes1 = fig.add_subplot(5, 1, 2)


axes1.set_ylabel('average')
axes1.plot(numpy.mean(data, axis=0), drawstyle='steps-pre')

axes2.set_ylabel('max')
axes2.plot(numpy.amax(data, axis=0),drawstyle="steps-mid")



fig.tight_layout()

matplotlib.pyplot.savefig('inflammation.png')
matplotlib.pyplot.show()
```
